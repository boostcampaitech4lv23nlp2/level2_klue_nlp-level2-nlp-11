## entity의 길이 확인하기

먼저 subject_entity와, object_entity의 type을 체크해야 했다.
왜냐하면, dataframe에서 봤을때는 dict(혹은 json)인 것 처럼 생겼지만, 사실 그냥 str이기 때문에 처리하는데 문제가 있었다.

따라서 str을 처리가능한 데이터 타입으로 변경해주어야 했다.

```python
type(df["subject_entity"][0])
```

```python
df["subject_entity"] = df["subject_entity"].apply(eval)
df["object_entity"] = df["object_entity"].apply(eval)
#object_entity와 subject_entity의 value를 dict로 변경.
type(df["subject_entity"][0])
```

```python
#object_entity(dict)를 unpacking하여 새로운 DF로 생성했다.
obj_df = pd.json_normalize(df["object_entity"], sep="_")
obj_df = obj_df.add_prefix("obj_")
obj_df
```

```python
#subj도 마찬가지로 새로운 DF를 만들었다.
subj_df = pd.json_normalize(df["subject_entity"], record_prefix="subj")
subj_df = subj_df.add_prefix("subj_")
subj_df
```

```python
#3개의 DF를 merge해주었다. 이로써 subj와 obj에 대한 분석이 가능해졌다.
df = pd.merge(df, subj_df, how="outer", left_index=True, right_index=True)
df = pd.merge(df, obj_df, how="outer", left_index=True, right_index=True)
df = df.drop(["subject_entity", "object_entity"], axis=1)
df
```
