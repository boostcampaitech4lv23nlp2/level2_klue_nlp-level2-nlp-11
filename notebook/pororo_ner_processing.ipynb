```python
import pandas as pd
import numpy as np
```

```python
train_df = pd.read_csv("/opt/ml/level2_klue_nlp-level2-nlp-11/dataset/train/train.csv")
```

```python
train_df
```

```python
per_other_family_df = pd.read_csv("./per:other_family.csv")
per_place_of_death = pd.read_csv("./per:place_of_death.csv")
per_place_of_residence = pd.read_csv("./per:place_of_residence.csv")
per_place_of_birth = pd.read_csv("./per:place_of_birth.csv")

concatenate_list = [per_other_family_df, per_place_of_death , per_place_of_residence , per_place_of_birth ]
print(f"per_other_family_df 의 개수 : {len(per_other_family_df)}")
print(f"per_place_of_death 의 개수 : {len(per_place_of_death)}")
print(f"per_place_of_residence 의 개수 : {len(per_place_of_residence)}")
print(f"per_place_of_birth 의 개수 : {len(per_place_of_birth)}")
```

```python
concat_df = pd.concat(concatenate_list, join="outer" )
concat_df
```

```python
print(f"concatenate_df 의 개수 : {len(concat_df)}")
```

```python
subjt = concat_df['subject_entity'].apply(lambda x: pd.Series(eval(x))).add_prefix('subj_')
objt = concat_df['object_entity'].apply(lambda x: pd.Series(eval(x))).add_prefix('obj_')
concat_df = pd.concat([concat_df[["sentence", "subject_entity", "object_entity", "label", "source"]], subjt, objt], axis=1)
```

```python
concat_df
```

```python
from functools import partial

def change_alpha(x, col1 , col2):
    sub_word, obj_word = x[col1] , x[col2]

    if (sub_word.isalpha() | obj_word.isalpha()) :
        return False
    else :
        return True


ff = partial(change_alpha , col1 = "subj_word", col2="obj_word")
```

```python
concat_df[['is_alpha']]= concat_df.apply(lambda x : ff(x),axis=1)
concat_df['is_alpha'].value_counts()
```

```python
remove_eng_word = (concat_df.is_alpha == False)
remove_eng_concat_df = concat_df.loc[remove_eng_word, :]
```

```python
remove_eng_concat_df
```

```python
remove_eng_concat_df = remove_eng_concat_df.drop(
    ["subj_start_idx", "subj_end_ids", "subj_type", "obj_start_idx", "obj_end_ids", "obj_type", "subj_word", "obj_word", "is_alpha"], axis=1)
```

```python
remove_eng_concat_df
```

```python
pororo_ner_df = pd.concat([train_df, remove_eng_concat_df], join="outer")
```

```python
#reset id , id type
pororo_ner_df['id'] = np.arange(len(pororo_ner_df))
pororo_ner_df = pororo_ner_df.astype({'id':'int64'})
```

```python
#추출전 데이터 shuffle
pororo_ner_df = pororo_ner_df.sample(frac=1)
pororo_ner_df
```

```python
pororo_ner_df.to_csv('./pororo_ner_df.csv',index=False, encoding = 'utf-8')
```
