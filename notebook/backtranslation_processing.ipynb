```python
import pandas as pd
```

```python
df = pd.read_csv("./backtranslate.csv")
df
```

```python
subjt = df['subject_entity'].apply(lambda x: pd.Series(eval(x))).add_prefix('subj_')
objt = df['object_entity'].apply(lambda x: pd.Series(eval(x))).add_prefix('obj_')
df = pd.concat([df[["new_sentence", "label", "source"]], subjt, objt], axis=1)
```

```python
def is_in_subj_and_obj(row):
    if row["subj_word"] in row["new_sentence"] and row["obj_word"] in row["new_sentence"]:
        return True
    return False
```

```python
m = df.apply(is_in_subj_and_obj, axis=1)
print(m)
```

```python
good_aug_df = df[m]
```

```python
good_aug_df
```

```python
def assign_index(row):
    subj_start = row["new_sentence"].find(row["subj_word"])
    obj_start = row["new_sentence"].find(row["obj_word"])
    return (subj_start, subj_start + len(row["subj_word"]), obj_start, obj_start + len(row["obj_word"]))
```

```python
good_aug_df[["subj_start_idx", "subj_end_idx", "obj_start_idx", "obj_end_idx"]]= good_aug_df.apply(lambda x: pd.Series(assign_index(x)), axis=1)
good_aug_df
```

```python
good_aug_df["subject_entity"] = good_aug_df.apply(
    lambda x:"{" + f"'word':'{x.subj_word}','start_idx':{x.subj_start_idx},'end_ids':{x.subj_end_idx},'type':'{x.subj_type}'" + "}", axis=1)
good_aug_df["object_entity"] = good_aug_df.apply(
    lambda x:"{" + f"'word':'{x.obj_word}','start_idx':{x.obj_start_idx},'end_ids':{x.obj_end_idx},'type':'{x.obj_type}'" + "}", axis=1)
```

```python
good_aug_df = good_aug_df.drop(["subj_start_idx", "subj_end_idx", "subj_type", "obj_start_idx", "obj_end_idx", "obj_type", "subj_word", "obj_word"], axis=1)
good_aug_df.rename(columns={'new_sentence':'sentence'})
```

```python
df = pd.read_csv("./backtranslate.csv")
df
```

```python
df = pd.concat([df, good_aug_df])
df
```

```python
df.to_csv('./concat_backtrans.csv', index=False)
```
